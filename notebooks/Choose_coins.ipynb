```python
import json

# Specify the path to your JSON file
json_file_path = '../data/all_coins_by_mc.json'

# Open the JSON file and load its contents
with open(json_file_path, 'r') as file:
    data = json.load(file)

# Now, 'data' contains the contents of the JSON file as a Python object (dictionary, list, etc.)
print(data)
```

```python
# Specify the path to your JSON file
json_file_path_1 = '../data/all_coins_by_mc_1.json'
json_file_path_2 = '../data/all_coins_by_mc_2.json'


# Open the JSON file and load its contents
with open(json_file_path_1, 'r') as file:
    data = json.load(file)

with open(json_file_path_2, 'r') as file:
    data = data + json.load(file)
# Now, 'data' contains the contents of the JSON file as a Python object (dictionary, list, etc.)
print(data)
```

```python
symbols_list = [stock["symbol"] for stock in data]
```

```python
symbols_list
```

```python
id_list = [stock["id"] for stock in data]
id_list
```

```python
import csv

# Example list

# Specify the path to your CSV file
csv_file_path = '../data/id_coins.csv'

# Open the CSV file in write mode
with open(csv_file_path, 'w', newline='') as file:
    # Create a CSV writer object
    csv_writer = csv.writer(file)

    # Write the list to the CSV file
    csv_writer.writerow(id_list)

print(f'The list has been saved to {csv_file_path}')
```
